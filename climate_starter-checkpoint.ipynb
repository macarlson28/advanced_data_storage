# Reflect Tables into SQLAlchemy ORM

```python
# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///Resources/hawaii.sqlite")
```

```python
# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)
```

```python
# We can view all of the classes that automap found
Base.classes.keys()
```

```python
# Save references to each table
Measurements = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

# Exploratory Climate Analysis

```python
# Design a query to retrieve the last 12 months of precipitation data and plot the results
#find the date and preceptiation from the measurements class
precipitation=session.query(Measurements.date, Measurements.prcp).filter(Measurements.date > '2016-08-23' ).order_by(Measurements.date).all()

len(precipitation)
```

```python
#find data to use in dataframe
precipitation_date = [i[0] for i in precipitation]
precipitation_prcp = [i[1] for i in precipitation]
len(precipitation_date)
```

```python
#put the results into a dataframe to be plotted
precipitation_df = pd.DataFrame(
    {'date': precipitation_date,
     'prcp': precipitation_prcp
    })
#drop all NAN values
precipitation_df.dropna(how='any')
```

```python
#plot the results
ax = precipitation_df.plot(kind='bar', title ="Precipitation", figsize=(15, 10), legend=True, fontsize=12)
ax.set_xlabel("Date", fontsize=12)
plt.xticks([])
plt.tight_layout()
plt.show()
```

```python
# This function called `calc_temps` will accept start date and end date in the format '%Y-%m-%d' 
# and return the minimum, average, and maximum temperatures for that range of dates
def calc_temps(start_date, end_date):
    """TMIN, TAVG, and TMAX for a list of dates.
    
    Args:
        start_date (string): A date string in the format %Y-%m-%d
        end_date (string): A date string in the format %Y-%m-%d
        
    Returns:
        TMIN, TAVE, and TMAX
    """

    return session.query(func.min(Measurements.tobs), func.avg(Measurements.tobs), func.max(Measurements.tobs)).\
        filter(Measurements.date >= start_date).filter(Measurements.date <= end_date).all()

# function usage example
print(calc_temps('2012-02-28', '2012-03-05'))
```
